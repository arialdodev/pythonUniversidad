#Algoritmo para hacer conteos

```python
numeros = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]


conteo = {}

for numero in numeros:
    if numero in conteo:
        conteo[numero] += 1
    else:
        conteo[numero] = 1

for numero, cantidad in conteo.items():
    print(f"El número {numero} aparece {cantidad} veces.")
```

```python
# importa collections y le pone de apodo Counter
from collections import Counter

# define una varible y se asigna de valor una cadena de tecto
texto = "El ordenamiento por conteo es un algoritmo de ordenamiento no basado en comparaciones . Es especialmente eficiente cuando el rango de valores de entrada es pequeño en comparación con el número de elementos a ordenar. La idea básica del ordenamiento por conteo es contar la frecuencia de cada elemento en la matriz de entrada y usar esa información para colocar los elementos en sus posiciones correctas."

# Divide el texto en una lista de palabras separadas por espacios.
palabras = texto.split()

# Usa Counter para contar cuántas veces aparece cada palabra en la lista.
frecuencia = Counter(palabras)

# Recorre cada palabra y su respectiva cantidad de veces que aparece.
print("Frecuencia de palabras:")
for palabra, conteo in frecuencia.items():
    print(f"{palabra}: {conteo}") # Muestra la palabra y cuántas veces aparece en el texto.
```

```python
# crea un una variable y el valor es un array con tispos de frutas
palabras = ["manzana", "To begin to toboggan first buy a toboggan, but don't buy too big a toboggan. Too big a toboggan is too big a toboggan to buy to begin to toboggan."]

# Diccionario vacío
contador = {}

# hace un bucle con las palabras de la lista osea que los recorre uno por uno
for palabra in palabras:
    # si la palabra ya esta en el diccionario aumenta uno mas al contador
    if palabra in contador:
        contador[palabra] += 1
    # Si la palabra no esta en el diccionario la agregamos con un contador de 1
    else:
        contador[palabra] = 1

# Imprimimos las palabras y su cantidad de repeticiones
print(contador)
```

```python
# 1. Definimos una función para contar todas las palabras del texto
def contar_palabras(texto):
    # 1.1 Separamos el texto en palabras usando split()
    palabras = texto.split()
    
    # 1.2 Contamos cuántas palabras hay usando len()
    total = len(palabras)
    
    # 1.3 Mostramos el total de palabras al usuario
    print("El texto tiene", total, "palabras")

# 2. Definimos una función para contar cuántas veces se repite cada palabra
def contar_repeticiones(texto):
    # 2.1 Convertimos el texto a minúsculas para que "Hola" y "hola" se cuenten igual
    # 2.2 Separamos el texto en una lista de palabras
    palabras = texto.lower().split()
    
    # 2.3 Creamos un diccionario vacío para guardar la cantidad de veces que aparece cada palabra
    contador = {}
    
    # 2.4 Recorremos cada palabra en la lista
    for palabra in palabras:
        # 2.4.1 Si la palabra ya está en el diccionario, le sumamos 1
        if palabra in contador:
            contador[palabra] += 1
        # 2.4.2 Si no está, la agregamos con valor 1
        else:
            contador[palabra] = 1
    
    # 2.5 Mostramos todas las palabras y cuántas veces se repite cada una
    for palabra, cantidad in contador.items():
        print(f"La palabra '{palabra}' se repite {cantidad} veces")

# 3. Parte principal del programa
# 3.1 Le pedimos al usuario que escriba un texto
texto = input("Ingresa un texto que tenga más de 45 palabras:\n")

# 3.2 Verificamos si el texto tiene más de 45 palabras
if len(texto.split()) > 45:
    # 3.2.1 Llamamos a la función para contar las palabras
    contar_palabras(texto)
    
    # 3.2.2 Llamamos a la función para mostrar repeticiones
    contar_repeticiones(texto)
else:
    # 3.3 Mostramos un mensaje si el texto tiene 45 palabras o menos
    print("El texto debe tener más de 45 palabras.")
```
